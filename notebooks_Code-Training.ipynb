 A data type describes the kind of value a variable holds. 
 For example, a number, a piece of text, or a list of items. Programming languages use data types so they know how to store and work with different kinds of information.

Python is a dynamically-typed language like JavaScript, meaning you don't need to explicitly declare types for variables. The language knows what data type a variable is based on what you assign to it.

Here are some examples:

name = 'John Doe' # Python knows this is a string
age = 25 # Python knows this is an integer

Here are the most common data types you'll use in Python:

Integer: A whole number w/o decimals, example: 10 or -5 

```python
my_integer_var = 10
print('Integer:', my_integer_var) # Integer: 10
```

Float: A number with a decimal point, like 4.41 or -0.4

```python
my_float_var = 4.50
print('This is my', 'float:', my_float_var) # This is my float 4.5
```

String: A sequence of characters (letters) enclosed in single or double quotation marks like, "Hello world" & 'blue' 

```python
my_string_var = 'hello to me'
print('string is:' , my_string_var, 'and i am cool') #string is: hello to me and i am cool
```

Boolean: A true or false type, written as True or False #notice the capital T and F

```python
my_boolean_var = True 
print( 'is the sky', 'blue?', my_boolean_var) #is the sky blue? True
```

*Important:{sets} are useful to automatically eliminate duplicate values

{sets} are unordered, meaning the  are not assigned to a position or index


Suppose you have a [list] of randomized numbers and you only want the unique values, simply convert the list to a set...which automatically filter out repeated values 

```python
#LIST TO SET EXAMPLE 
numbers = [1,1,1,5,4,3,2,6,8,7,9,9,9,9,8]
set(numbers)
```

then if you want to work with the list again, simply cast it back into the list:

```python
#LIST TO SET, THEN BACK TO LIST EXAMPLE 
numbers = [1,1,1,5,4,3,2,6,8,7,9,9,9,9,8]
set(numbers)
list(set(numbers))
```

[list] and (tuple)'s have ordered collections, meaning each value has a position called an index, assigning the [list] and (tuple) to a variable

The index is used to access any of the values position individually  

```python
#LIST
my_list = ['a', 'b', 'c', 'a']
print(my_list[0])
```

```python
#TUPLE INDEXING 
my_tuple = ('a', 'b', 'c', 'a')
print(my_tuple[2]) #to get the correct index value you must use [list] bracket around the number position
```
